```python
#The goal of this project is to build and evaluate a regression model to predict using the following self created dataset.
```

```python
#Import necessary libraries
import pandas as pd #for data manipulation and analysis
import numpy as np #for numerical operations
import sklearn
from sklearn.model_selection import train_test_split #to split the dataset
from sklearn.linear_model import LinearRegression #a simple regression model
from sklearn.metrics import root_mean_squared_error, r2_score #for model evaluation
import matplotlib.pyplot as plt #for data visualization
import seaborn as sns #for enhanced visualization
```

```python
#Load the dataset
```

```python
data = {
    'Average Daily Usage (hours)': [3, 4, 6, 7, 8, 10, 12],
    'Monthly Bill ($)': [50, 60, 80, 95, 110, 130, 150]
}
df = pd.DataFrame(data)
print(df)
```

```python
#explore the dataset
```

```python
print("Dataset Shape:", df.shape) #Print the dimensions of the dataset
```

```python
df.describe()
```

```python
#Data Cleaning
```

```python
#check for missing values
print("Missing Values:", df.isnull().sum())
```

```python
#feature Selection
```

```python
#selecting features that are likely to affect house prices
```

```python
X=df[['Average Daily Usage (hours)']] #Feature
y=df['Monthly Bill ($)'] #target variable 
```

```python
#split the dataset into training and testing sets
#test_size=0.2 means 20% of the data will be used for testing
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Model training
```

```python
#Initialize the linear regression model
model=LinearRegression()
#Fit the model on the training data
model.fit(X_train,y_train)
```

```python
#Model Evaluation
#Make predictions on the test data
y_pred=model.predict(X_test)
```

```python
#calculate evaluation metrics
mse=root_mean_squared_error(y_test,y_pred) #Mean Squared Error
r2=r2_score(y_test,y_pred) #R-squared score
```

```python
#Print evaluation metrics
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```

```python
#Visualize the results
plt.figure(figsize=(10,6))
sns.scatterplot(x=y_test,y=y_pred) #scatterplot of actual vs predicted prices
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs. Predicted House Prices')
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='red') #diagonal line for perfect predictions
plt.show()
```

```python
# Create the new input for prediction with proper column name
new_input = pd.DataFrame([[9]], columns=['Average Daily Usage (hours)'])

# Predict the monthly bill for an average daily usage of 5 hours
predicted_bill = model.predict(new_input)

# Display the prediction
print(f"Predicted Monthly Bill for Average Daily Usage = 9 hours: ${predicted_bill[0]:.2f}")
```
